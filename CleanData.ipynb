# 補遺漏日

```python
ori_data = pd.read_csv('covid_全區.csv')  # 讀取訓練集
ori_data.drop('id',axis=1,inplace=True)
ori_data.drop('縣市別',axis=1,inplace=True)
ori_data.drop('區域',axis=1,inplace=True)
data=ori_data
data=data.reindex(index=data.index[::-1])
data.to_csv('tmp.csv',index=False)
data.reset_index(inplace=True,drop=True)
data
```

```python
mon=[31,28,31,30,31,30,31,31,30,31,30]


# data.index[data['個案公佈日']=='2022-01-20'].tolist()
data[data['個案公佈日']=='2022-01-20'].index.values
```

```python
date=[]
num=[]
for i in range(len(mon)):
  for j in range(mon[i]):
    _date='2022-{0}-{1}'.format(str(i+1).zfill(2),str(j+1).zfill(2))
    if _date in data.values:
      index=data.index[data['個案公佈日']==_date].tolist()[0]
      _num=data.iat[index,1]
    else:
      _num = 0
    date.append(_date)
    num.append(_num)


end_data = pd.DataFrame(columns=['日期','人數'])
end_data['日期']=date
end_data['人數']=num

end_data
```

```python
print(end_data['人數'].shape[0])
for i in range(1,end_data['人數'].shape[0]):
  if end_data['人數'][i] == 0:
    j=i+1
    while(end_data['人數'][j] == 0):
      j+=1
    end_data['人數'][i]= (end_data['人數'][i-1] + end_data['人數'][j]) //2

end_data.to_csv('lstm_data2.csv',index=False)
```
